```python
l=5
m=10
n=20
o=20
p=30
i=0
while True:
    print("                ESSENCIAL ITEMS                    ")
    print("%10s %10s %10s %10s"%("s.no","item","quantity","cost/item"))
    print("%10d %10s %10d %10f"%(1,"Biscuits",l,20.50))
    print("%10d %10s %10d %10f"%(2,"Cereals",m,90.00))
    print("%10d %10s %10d %10f"%(3,"Chicken",n,200.00))
    print("%10d %10s %10d %10f"%(4,"Oats",o,39.99))
    print("%10d %10s %10d %10f"%(5,"Rice",p,79.99))


        
    bis=int(input("how many biscuits packets are required: "))
    cer=int(input("how many Cereal packets are required: "))
    chi=int(input("how many Chi packets are required: "))
    oat=int(input("how many Oats packets are required: "))
    ric=int(input("how many Rice packets are required: "))
    
    if l==0 or m==0 or n==0 or o==0 or p==0:
            if l==0 and bis!=0:
                bis=0
                print("Biscuits not available:" )
            if m==0 and cer!=0: 
                cer=0
                print("Cereals  not available:" )
            if n==0 and chi!=0:
                chi=0
                print("Chicken not available:" )
            if o==0 and oat!=0:
                oat=0
                print("Oats not available:" )
            if p==0 and ric!=0:
                ric=0
                print("Rice not available:" )
        
        
    if bis>l or cer>m or chi>n or oat>o or ric>p:
            if bis>l:
                print("extra product entered for biscuits")
                bis=int(input("enter new value for biscuit:"))
            if cer>m:
                print("extra product entered for cereals")
                cer=int(input("enter new value for cereals:"))
            if chi>n:
                print("extra product enterd for chicken")
                chi=int(input("enter new value for chicken:"))
            if oat>o:
                print("extra product entered for oats")
                oat=int(input("enter new value for oats:"))
            if ric>p:
                print("extra product entered for rice")
                ric=int(input("enter new value for rice:"))
                
                
            

    print("    Fill in your delivery details   ")
    name=input("enter your name:")
    address=input("enter your address")
    dist=int(input("distance from the store 5/10/15 km:"))
    del_chr=dist*2
    print("Delivery charge for ",dist,"km is ",del_chr)

    def bill():
            
        a=bis*20.50
        b=cer*90.00
        c=chi*200.00
        d=oat*39.99
        e=ric*79.99
        print("                BILL                ")
        print("%10s %10s %10s %10s"%("S.No","Item","Quantity","Toatal Cost"))
        print("%10d %10s %10d %10f"%(1,"Biscuits",bis,a))
        print("%10d %10s %10d %10f"%(2,"Cereals",cer,b))
        print("%10d %10s %10d %10f"%(3,"Chicken",chi,c))
        print("%10d %10s %10d %10f"%(4,"Oats",oat,d))
        print("%10d %10s %10d %10f"%(5,"Rice",ric,e))
        
        global l
        global m
        global n
        global o
        global p
        l=l-bis
        m=m-cer
        n=n-chi
        o=o-oat
        p=p-ric
        total_item_cost=0
        total=0
        total_item_cost=a+b+c+d+e 
    
        print("Total_Item_cost:",total_item_cost)
        total=total_item_cost+del_chr
        print("Total_bill_amount = Total_item_cost +delivery_charges  :  ",total)
        global i
        i+=1
        print("         ADDRESS DETAILS     ")
        print("Name : ",name)
        print("Address : ",address)
        sum_total=0
        for j in range(i):
            sum_total=sum_total+total
            print("the new total is",sum_total)
            j+=1
    bill()   
    shop=input("Do you want to continue shopping y/n  :  ")
    if shop=='y':
        continue
    if shop=='n':
        break
    
print("Thank You!")
```
